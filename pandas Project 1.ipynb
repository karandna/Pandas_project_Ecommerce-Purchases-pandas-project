```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
df.head

parent_themes = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
parent_themes.head()
```

```python
merged2=parent_themes.merge(df,left_on='name',right_on='theme_name')
merged2.head()
```

```python
#### lets join the parent theme  from df and name theme from parent_themes,lef

merged=df.merge(parent_themes,left_on='parent_theme',right_on='name')
merged.drop(columns='name_y',inplace=True)
merged.head()
```

```python
###take true values from is_licensed 
licensed=merged[merged['is_licensed']]
#### drop all null values froms licensed
licensed=licensed.dropna()
licensed.head()
```

```python

star_wars=licensed[licensed['parent_theme']=='Star Wars']
the_force=float(star_wars.shape[0]/licensed.shape[0])
print(the_force)
star_wars.shape[0] 
```

```python
###gropby
licensed_sorted=licensed.sort_values('year')
licensed_sorted['count']=1
summed_df=licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.drop(['set_num','name_x'],axis=1)
```

```python
 ## get only max rows in a groupby 
max_df=summed_df.sort_values('count',ascending=False).drop_duplicates(['year'])
max_df.drop(['set_num','name_x','theme_name'],axis=1)
```
